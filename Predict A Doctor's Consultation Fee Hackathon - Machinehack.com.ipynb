```python
# Feature Engineering
```

```python
# splitting the Place into two diff columns, They are delimited by ','
df['Area']=df.Place.str.split(',').str[0]
df['City']=df.Place.str.split(',').str[1]
```

```python
# replacing null values in the new columns with 'others'
df.Area.fillna("others",inplace=True)
df.City.fillna("others",inplace=True)
```

```python
# splitting qualification into diff columns. They are delimited by ','
df['qual1']=df.Qualification.str.split(',').str[0]
df['qual2']=df.Qualification.str.split(',').str[1]
df['qual3']=df.Qualification.str.split(',').str[2]
```

```python
# replcaing null values in the new qual(1,2,3) coulumns
df.qual1.fillna("XXX",inplace=True)
df.qual2.fillna("XXX",inplace=True)
df.qual3.fillna("XXX",inplace=True)
```

```python
# splitting Experience column in to int value of experience
df['exp_years']=df.Experience.str.split(' ').str[0].astype(int)
```

```python
# Null of ratings are filled with 0%
df.Rating.fillna('0%',inplace=True)
df['Rating_%']=df.Rating.str.split("%").str[0]
df['Rating_%']=df['Rating_%'].astype(int)
```

```python
# feature engineering the Misc column, we take the rating as misc1 and no. of feedback as misc3

df['Misc1']=df.Miscellaneous_Info.str.split('%').str[0]
df['Misc1_len']=df.Misc1.str.len()
df.loc[df.Misc1_len >3,['Misc1']]=0
df.Misc1.fillna(0,inplace=True)
df.Misc1=df.Misc1.astype(int)
df['Misc2']=df.Miscellaneous_Info.str.split('% ').str[1]
df['Misc3']=df.Misc2.str.split(' ').str[0]
df['Misc3_len']=df.Misc3.str.len()
df.loc[df.Misc3_len >3,['Misc3']]=0
df.Misc3.fillna(0,inplace=True)
df.loc[df.Misc3 == ',',['Misc3']]=0
df.Misc3=df.Misc3.astype(int)
```

```python
#df.Misc3=np.where((df['Rating_%'] > 0) & (df.Misc3 == 0),1,df.Misc3)
df.loc[(df['Rating_%'] > 0) & (df.Misc3 == 0),['Misc3']]=1
```

```python
df.head()
```

```python
# Feature Selection

df.columns
```

```python
df = df.drop(df[['Experience','Misc1','Misc2','Miscellaneous_Info','Place','Qualification','Rating','Misc1_len','Misc3_len']], axis=1)
```

```python
df=df.sort_values(by='Fees',ascending=False).reset_index(drop=True)
```

```python
# Encoding Categorical variables
from sklearn.preprocessing import LabelEncoder
cat_cols=['Profile','Area','City','qual1','qual2','qual3']
for i in cat_cols:
    l=LabelEncoder()
    df[i]=l.fit_transform(df[i])
df.head()
```

```python
df_train = df[0:5961]
df_test = df[5961:]
df_test = df_test.drop(['Fees'], axis =1)
```

```python
# creating x and y
df_xgb = df_train[['qual1','qual2','exp_years', 'Rating_%','Profile','Area','City','Misc3','Fees']]
X = df_xgb.drop(['Fees'], axis=1)
y = df_xgb.Fees
```

```python
# Importing XGBoost
import xgboost as xgb
from sklearn.metrics import mean_squared_error
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
# Cross-Validation
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1, max_depth = 5, n_estimators = 100)

xg_reg.fit(X_train,y_train)
preds = xg_reg.predict(X_test)
```

```python
# Error & Predictions
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))

df_test_xgb = df_test[['qual1','qual2','exp_years', 'Rating_%','Profile','Area','City','Misc3']]

preds_1 = xg_reg.predict(df_test_xgb)
df_test_xgb['Fees'] = preds_1
df_test_xgb.to_csv('Submission1.csv')
```
